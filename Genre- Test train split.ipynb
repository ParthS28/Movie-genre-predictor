```python
import numpy as np
import pandas as pd
data = pd.read_csv('movie_poster.csv')
data
```

```python
df = data.drop(['tid', 'image_url', 'url', 'year'], axis = 1)
df['genre'] = pd.Series(np.random.randn(len(df)), index=df.index)
for i in range(len(df)):
    if df.iloc[i]['action'] == True:
        df.at[i, 'genre'] = 0
    elif df.iloc[i]['scifi'] == True:
        df.at[i, 'genre'] = 1
    elif df.iloc[i]['drama'] == True:
        df.at[i, 'genre'] = 2
    elif df.iloc[i]['romance'] == True:
        df.at[i, 'genre'] = 3

df
```

```python
y = df['genre']
import tensorflow as tf
y_categorical = tf.keras.utils.to_categorical(y)
y_categorical
```

```python
tid = data['tid']
import matplotlib.pyplot as plt
import os
import cv2
filename = 'posters/' + tid[0] + '.jpg'
for i in range(len(tid)):
    path = 'posters/' + tid[i] + '.jpg'
    img_array = cv2.imread(path)
    plt.imshow(img_array)
    plt.show()
    break
```

```python
size = 150
path = 'posters/' + tid[1] + '.jpg'
img_array = cv2.imread(path)
new_img = cv2.resize(img_array, (size, size))
plt.imshow(new_img)
plt.show()
```

```python
dataset = []

def create_training_data():
    for i in range(len(tid)):
        #try:
        path = 'posters/' + tid[i] + '.jpg'
        img_array = cv2.imread(path)
        new_img = cv2.resize(img_array, (size, size))
        dataset.append(new_img)
        print(i)
        #except Exception as e:
            #pass

create_training_data()
print(len(dataset))
```

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset, y_categorical, test_size = 0.2, random_state = 4)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization

model = Sequential()
model.add(Conv2D(16, kernel_size = (3, 3), activation = 'relu', input_shape = (size, size, 3)))
#model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(256, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(4, activation='sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
x = np.asarray(dataset)
x = x.reshape((2438, size, size, 3))
x_train, x_test, y_train, y_test = train_test_split(x, y_categorical, test_size = 0.2, random_state = 4)
model.fit(x_train, y_train, epochs = 10, batch_size = 16, verbose = 1, validation_split = 0.2)
```

```python
predict = model.predict(x_test)
predict
```

```python
pred = []
for i in range(len(predict)):
    max = 0
    a = predict[i]
    for j in range(4):
        if a[max] < a[j]:
            max = j
    for j in range(4):
        if j == max:
            a[j] = 1
        else:
            a[j] = 0
    pred.append(a)

count = 0
for i in range(len(pred)):
    a = pred[i]
    b = y_test[i]
    for j in range(4):
        if a[j] == 1 and b[j] == 1:
            count += 1;
count/len(pred)
```
